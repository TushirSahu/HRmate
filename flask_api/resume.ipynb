```python
import json
 
data = [json.loads(line) for line in open('Entity Recognition in Resumes.json', 'r')]
contentlist = []
namelist = []
locationlist = []
emaillist = []
degreelist = []
skilllist = []


print(data)
for line in data:
    contentlist.append(line["content"]) 
newlist = []
for line in data:
    newlist.append(line["annotation"])

def get_annotation(newlist,xlist,label):
    
    for i in range(len(newlist)):
        for dict1 in newlist[i]:
            if dict1['label'] == [label] and len(xlist) == i:
                xlist.append(dict1['points'][0]["text"])
                break
            elif len(xlist) != i:
                xlist.append(None)
            elif len(xlist) == len(newlist) - 1 :
                xlist.append(None)
                break
            else:
                continue
    return xlist
    
      
emaillist = get_annotation(newlist,emaillist,"Email Address")
skilllist=get_annotation(newlist,skilllist,"Skills")
namelist = get_annotation(newlist,namelist, "Name")
locationlist = get_annotation(newlist,locationlist,"Location")
degreelist = get_annotation(newlist,degreelist, "Degree")
        
```

```python
import spacy
import re
def get_location(resume):
    pattern = r'([a-zA-Z\,\s\/\-]+)(?=Email)'
    text = re.search(pattern, resume.replace('\n',' '))
    if text == None:
        return None

    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text.group().replace('-',' '))
    for ent in doc.ents:
        if ent.label_ == 'GPE':
            return ent.text
        
df["location"] = df["resume"].apply(lambda x: get_location(x))
df[["name","location"]].sample(10)
```

```python
def get_edu_info(resume):
    pattern =r'(?<=EDUCATION|Education)([A-Za-z0-9\-\s\',.]*)(?=SKILLS|Skills)'
    matched = re.search(pattern, resume)
    if matched != None:
        return matched.group().replace('\n',' ')
    return None
```

```python
def recommend_resume(description, skills, location):
        df_location = df[df['location'] == location]

        cv_desc = CountVectorizer(stop_words='english')
        cv_skills = CountVectorizer(stop_words='english')
        cv_loc=CountVectorizer(stop_words='english')


        desc_vectors = cv_desc.fit_transform(df_location['resume'])
        skills_vectors = cv_skills.fit_transform(df_location['skills'])
        loca_vectors=cv_loc.fit_transform(df_location['location'])


        similarity_desc = cosine_similarity(desc_vectors)
        similarity_skills = cosine_similarity(skills_vectors)
        similarity_loc=cosine_similarity(loca_vectors)


        matching_indexes = []
        for i, skills_set in enumerate(df_location['skills']):
            if any(skill in skills for skill in skills_set.split(' ')):
                matching_indexes.append(i)


        matching_desc_indexes = []
        for i, desc in enumerate(df_location['resume']):
            if any(word in description for word in desc.split(' ')):
                matching_desc_indexes.append(i)


        final_matching_indexes = list(set(matching_indexes) & set(matching_desc_indexes))

        highest_similarity = 0
        matching_resume_index = 0
        for index in final_matching_indexes:
            overall_similarity = (similarity_desc[index].mean() +similarity_loc[index].mean()+ similarity_skills[index].mean())/3
            if overall_similarity > highest_similarity:
                highest_similarity = overall_similarity
                matching_resume_index = index
                df_location.loc[index, 'overall_similarity'] = overall_similarity

        
        pickle.dump(similarity_desc,open('description.pkl','wb'))
        pickle.dump(similarity_skills,open('skills.pkl','wb'))
        pickle.dump(similarity_loc,open('location.pkl','wb'))
        return df_location.iloc[final_matching_indexes].sort_values(by='overall_similarity',ascending=False)['name'].head(5)

```
